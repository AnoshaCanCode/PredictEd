```python
from scipy import stats
#Hours_Studied
#z-scores
df['Z_score'] = stats.zscore(df['Hours_Studied'])

#set threshold
outliers = df[df['Z_score'].abs() > 3]

outliers
```

```python
import numpy as np
#for skewed data, use median else use mean

z_scores = np.abs(stats.zscore(df['Hours_Studied']))

threshold = 3
# 5. Calculate the imputation value (mean of non-outliers)
mean_value = int(df['Hours_Studied'].mean())

df['Hours_Studied'] = np.where(z_scores > threshold, mean_value, df['Hours_Studied'])

print(f"\nMean of non-outliers used for imputation: {mean_value}")

print("\nDataFrame after imputation:")
print(df[['Hours_Studied']])
```

```python
#Exam_Score
#IQR
Q1 = df['Exam_Score'].quantile(0.25)
Q3 = df['Exam_Score'].quantile(0.75)
IQR = Q3-Q1

lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR

outliers = df[(df['Exam_Score'] < lb) | (df['Exam_Score'] > ub)]

outliers
```

```python
z_scores = np.abs(stats.zscore(df['Exam_Score']))

threshold = 3
# 5. Calculate the imputation value (median of non-outliers)
median_value = int(df['Exam_Score'].median())

df['Exam_Score'] = np.where(z_scores > threshold, median_value, df['Exam_Score'])

print(f"\nMedian of non-outliers used for imputation: {median_value}")

print("\nDataFrame after imputation:")
print(df[['Exam_Score']])
```

```python
#for binary attributes
pd.set_option('future.no_silent_downcasting', True)
from sklearn.preprocessing import LabelEncoder

df['Extracurricular_Activities'] = df['Extracurricular_Activities'].replace({'Yes': 1, 'No': 0})

df['Internet_Access'] = df['Internet_Access'].replace({'Yes': 1, 'No': 0})
df['Learning_Disabilities'] = df['Learning_Disabilities'].replace({'Yes': 1, 'No': 0})
```

```python
#for ordinal attributes

# Define the order manually
size_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['Parental_Involvement'] = df['Parental_Involvement'].replace(size_map)

df['Access_to_Resources'] = df['Access_to_Resources'].replace(size_map)

df['Motivation_Level'] = df['Motivation_Level'].replace(size_map)

df['Family_Income'] = df['Family_Income'].replace(size_map)

df['Teacher_Quality'] = df['Teacher_Quality'].replace(size_map)

size_map2 = {'Negative': 1, 'Neutral': 2, 'Positive': 3}
df['Peer_Influence'] = df['Peer_Influence'].replace(size_map2)

size_map3 = {'Near': 1, 'Moderate': 2, 'Far': 3}
df['Distance_from_Home'] = df['Distance_from_Home'].replace(size_map3)
```

```python
#for nominal attributes
try:
    df = pd.get_dummies(df, columns=['School_Type', 'Parental_Education_Level', 'Gender'], prefix=['School_Type', 'PEL', 'Gender'], drop_first=False, dtype=int)      
#Ensures results are 1/0, not True/False
except KeyError:
    pass # This ignores the error silently
```
